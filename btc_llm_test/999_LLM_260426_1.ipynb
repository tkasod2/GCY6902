```python
import requests
import json

model_output = {
    "prob_up": 0.71,
    "pred_return": 0.008,
    "signal": "buy"
}

prompt = f"""
다음은 비트코인 예측모델 결과다:

{json.dumps(model_output, indent=2)}

이 결과를 기반으로
1) 매수 판단
2) 근거
3) 리스크
를 한국어로 설명해줘.
"""

payload = {
    "model": "qwen2.5:3b",
    "prompt": prompt,
    "stream": False,
    "options": {
        "num_predict": 500,
        "temperature": 0.7
    }
}

res = requests.post("http://localhost:11434/api/generate", json=payload)

print(res.json()["response"])
```
